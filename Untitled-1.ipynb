```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# 滤波器设计
def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = signal.butter(order, [low, high], btype='band')
    filtered_signal = signal.filtfilt(b, a, data)
    return filtered_signal

# Hill 肌肉模型参数
max_force = 1000  # 最大肌肉力
lo = 1.0  # 最优肌肉长度
delta_l = 0.15  # 长度变化参数
k = 0.5  # 刚度系数
v_max = 1.2  # 最大收缩速度
a = 0.25  # 速度参数

def hill_model(contraction_speed, activation_level):
    """
    根据 Hill 肌肉模型计算肌肉力与长度的关系
    """
    # 计算肌肉长度
    muscle_length = lo * (1 + contraction_speed / 0.45)
    
    # 力-长度关系
    f_length = np.exp(-((muscle_length - lo) / delta_l) ** 2)
    
    # 力-速度关系
    f_velocity = (v_max - contraction_speed) / (v_max + a * contraction_speed)
    
    # 计算肌肉力
    muscle_force = max_force * activation_level * f_length * f_velocity
    
    return muscle_length, muscle_force

# 特征提取
def extract_features(emg_signal):
    # 计算均方根值（RMS）
    rms = np.sqrt(np.mean(emg_signal**2))
    
    # 寻找极值点
    peaks, _ = signal.find_peaks(emg_signal)
    
    # 计算极值点之间的时间间隔（收缩速度）
    if len(peaks) >= 2:
        time_gaps = np.diff(peaks) / 1000  # 将时间间隔转换为秒
        contraction_speed = np.mean(time_gaps)
    else:
        contraction_speed = np.nan  # 如果没有足够的极值点，则返回 NaN
    
    return rms, contraction_speed

# 示例数据（这里使用随机数据作为示例）
np.random.seed(42)
fs = 1000  # 采样率
lowcut = 20.0  # 滤波器低截止频率
highcut = 450.0  # 滤波器高截止频率
rectus_femoris = np.random.randn(5000)  # 原始EMG数据
emg_normalized_rectus = np.abs(rectus_femoris) / np.max(np.abs(rectus_femoris))  # 归一化EMG数据

# 滤波处理
filtered_rectus = bandpass_filter(rectus_femoris, lowcut, highcut, fs)

# 设置时间间隔（200毫秒）
window_size = int(0.2 * fs)  # 200毫秒

# 初始化结果列表
activation_level_rectus = []
contraction_speed_rectus = []
muscle_length_rectus = []
muscle_force_rectus = []

# 遍历EMG信号，每隔200毫秒提取一次特征并计算Hill模型
for i in range(0, len(filtered_rectus), window_size):
    window_rectus = filtered_rectus[i:i+window_size]
    window_norm_rectus = emg_normalized_rectus[i:i+window_size]
    
    act_level_rectus, _ = extract_features(window_norm_rectus)
    _, cont_speed_rectus = extract_features(window_rectus)

    activation_level_rectus.append(act_level_rectus)
    contraction_speed_rectus.append(cont_speed_rectus)
    muscle_length, muscle_force = hill_model(cont_speed_rectus, act_level_rectus)
    muscle_length_rectus.append(muscle_length)
    muscle_force_rectus.append(muscle_force)

# 打印结果
print("Activation Level rectus (RMS):", activation_level_rectus)
print("Muscle Contraction Speed rectus(mean time gap between peaks):", contraction_speed_rectus)
print("Muscle Length rectus:", muscle_length_rectus)
print("Muscle Force rectus:", muscle_force_rectus)

# 可视化肌肉力与长度关系

plt.figure(figsize=(12, 8))



plt.subplot(2, 1, 1)
plt.plot(muscle_length_rectus, muscle_force_rectus, label='Rectus Femoris')
plt.xlabel('Muscle Length (normalized)')
plt.ylabel('Muscle Force (N)')
plt.title('Muscle Force-Length Relationship')
plt.legend()


plt.subplot(2, 1, 1)
plt.plot(muscle_length_rectus, muscle_force_rectus, label='Rectus Femoris')
plt.xlabel('Muscle Length (normalized)')
plt.ylabel('Muscle Force (N)')
plt.title('Muscle Force-Length Relationship')
plt.legend()

# 可视化肌肉力随时间变化
time_points = np.arange(0, len(muscle_force_rectus) * window_size, window_size) / 1000  # 时间点（秒）
plt.subplot(2, 1, 2)
plt.plot(time_points, muscle_force_rectus, label='Rectus Femoris')
plt.xlabel('Time (s)')
plt.ylabel('Muscle Force (N)')
plt.title('Muscle Force-Time Curve')
plt.legend()

plt.tight_layout()
plt.show()
```
